Любой физический признак определяется одним или несколькими генами. Каждый ген может существовать в нескольких различных формах, которые называются аллелями Число организмов в данной популяции, несущих определенный аллель, определяет частоту данного аллеля.

Например, у человека частота доминантного аллеля, определяющего нормальную пигментацию кожи, волос и глаз, равна 99%. Рецессивный аллель, детерминирующий отсутствие пигментации – так называемый альбинизм, - встречается с частотой 1%.

```python
import matplotlib.pyplot as plt
import numpy as np
import math as math
```

Мутации гена от состояния дикого типа к новому состоянию называют прямыми, а от мутантного к дикому — обратными; сам процесс обратного перехода называют реверсией гена. Прямые мутации чаще являются рецессивными, а обратные — доминантными. Исходный ген мутирует без промежуточных ступеней в новое состояние и, соответственно, обратно. У разных генов прямые и обратные мутации могут возникать с разной частотой

 Попробуем проанализировать, как меняется частота аллеля при наличии прямых и обратных мутаций.

p - частота аллеля в популяции (не важно, доминантного или рецессивного

u - частота прямой мутации

v - частота обратной мутации


```python
values = []
generations = list(range (1,201))
```

```python
p = 0.9 
u = 0.01
v = u
```

в данном случае примем, что частоты прямой и обратной мутации совпадают


**Изменение частоты аллеля в поколении $\Delta p$**

$ p' = p(1- \mu) + (1-p) \cdot \nu$

```python
def mutations (p,u,v):
  p1 = p*(1-u)+(1-p)*v
  return(p1)
```

```python
i=1
values.append(p)
while i < 200:
    p1=mutations(p,u,v)
    values.append(p1)
    p=p1
    i=i+1
```

```python
#Посмотрите на полученные значения
print(values)
```

```python
plt.figure(figsize=(10,10))
plt.plot(generations, values, linewidth=2.0)
plt.xlabel('Generation')
plt.ylabel('Allele frequency')
plt.show()
```

Попробуем посмотреть, как изменяется частота аллеля в зависимости от того, насколько низкая частота мутаций

```python
values_for_mutation_coefficient = [0.01, 0.5, 0.08]
results = {}
```

```python
for j in values_for_mutation_coefficient:
  u=j
  v=u/2
  p=0.5
  results[j] = []
  i=1
  results[j].append(p)
  while i < 200:
    p1=mutations(p,u,v)
    results[j].append(p1)
    p=p1
    i=i+1
```

берем здесь v = u/2, чтобы частота прямой и обратной мутации не равнялась, и мы увидели разницу


```python
plt.figure(figsize=(10,10))
plt.xlabel('Generation')
plt.ylabel('Allele frequency')
for i in values_for_mutation_coefficient:
  plt.plot(generations, results[i])
  plt.text(generations[-175], results[i][-175], f'u = {i}')
```
